```python
import os
import json
import random
import dask.bag as db
import re
from nltk.corpus import stopwords
import nltk

# scaricare roba
#nltk.download('stopwords')
```

```python
paths = []
for subdir in ('pmc_json/', 'pdf_json/'):
    with os.scandir('./data/document_parses/' + subdir) as dir:
        for entry in dir:
            if entry.name.endswith('.json'): paths.append(entry.path)
#print(paths[1])
sample = random.sample(paths, 5000)
bag = db.from_sequence(sample).map(lambda path: json.load(open(path)))
```

```python
stop_words = set(stopwords.words('english'))

def extractor(paper):
    field = paper.get("body_text", [])
    text = " ".join(f.get("text", "") for f in field)
    words = re.findall(r'\b\w+\b', text.lower())
    # questo è un'alternativa a nltk tokenize o text.split perché toglie direttamente anche la punteggiatura. \b indica il bound della parola, 
    # mentre \w indica la parola che può essere fatta da tutti i caratteri alfanum a-zA-Z0-9_ <- incluso underscore, + significa che prende più
    # caratteri consecutivi
    clean_words = [w for w in words if w not in stop_words]
    return clean_words
```

```python
words = bag.map(extractor)
words.take(2)
```
